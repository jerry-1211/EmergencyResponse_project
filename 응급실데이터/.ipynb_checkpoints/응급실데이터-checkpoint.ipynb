# 시도 구분

```python
df = pd.read_csv("시도정보.csv")
df.head()
```

```python
tmp = pd.read_excel("시도코드_법정동기준.xlsx")
tmp.head()
```

```python
# tmp.rename(columns={"시도코드_법정동":"CTPRVN_CD"},inplace=True)
```

```python
tmp["시도코드_법정동"] = tmp["시도코드_법정동"].astype(str) 
```

```python
tmp["시도코드_법정동"] = tmp["시도코드_법정동"].apply(lambda x : x[:2])
```

```python
tmp
```

```python
df["CTPRVN_CD"] = df["CTPRVN_CD"].astype(str)
```

```python
df = pd.merge(df,tmp,left_on="CTPRVN_CD", right_on="시도코드_법정동",how="left")
```

```python
df
```

```python
url = 'http://apis.data.go.kr/B552657/ErmctInfoInqireService/getEmrrmRltmUsefulSckbdInfoInqire'

city_name = []
town_name = []
hospital_name = []

for i in range(len(new_df)) :
    params ={'serviceKey' : decoding,
             'STAGE1' : new_df.iloc[i,0],
             'STAGE2' : new_df.iloc[i,1],
             'pageNo' : '1',
             'numOfRows' : '10' }

    response = requests.get(url, params=params)
    dic = xmltodict.parse(response.content)
    
    city = new_df.iloc[i,0]
    town = new_df.iloc[i,1]

    
    print(f"{i+1}. {city} {town}")  # 시도명 출력    
    
    try :
        # item의 개수가 1개 이상인 경우 
        items = dic["response"]["body"]["items"]["item"]

        for j in range(len(items)):            
            print("\t"+items[j]["dutyName"])
            hospital_name.append(items[j]["dutyName"])
            
            city_name.append(city)
            town_name.append(town)

    except :
        try:
            # item의 개수가 1개인 경우
            item = dic["response"]["body"]["items"]["item"]
            print("\t"+item["dutyName"])
            hospital_name.append(item["dutyName"])
            
            city_name.append(city)
            town_name.append(town)
            
        except : 
            # item의 개수가 0개인 경우
            hospital_name.append(None)
            print("\t NaN")
            
            city_name.append(city)
            town_name.append(town)
            
```

```python
df = pd.DataFrame([city_name,town_name,hospital_name],index=["city","town","hospital"]).transpose()
df
```

```python
df[df["hospital"].isna()]
```

```python
df.dropna(inplace=True)
```

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach",True)
service = Service(executable_path=ChromeDriverManager().install())  # Chrome driver 자동 업데이트
```

```python
browser = webdriver.Chrome(service=service, options=chrome_options)

browser.get("https://map.naver.com")
browser.implicitly_wait(10)

for key in ["거창적십자병원","국립중앙의료원"]:
    search = browser.find_element(By.CSS_SELECTOR,"input.input_search")
    
    time.sleep(2)
    search.send_keys(key)
    time.sleep(1)
    
    
    search.send_keys(Keys.ENTER)
    time.sleep(2)
    
    browser.switch_to.frame('searchIframe')
    browser.find_element(By.CSS_SELECTOR,"#_pcmap_list_scroll_container").click()
    time.sleep(1)

    browser.find_element(By.CSS_SELECTOR,"#_pcmap_list_scroll_container>ul>li:nth-of-type(1)>.qbGlu").click()
    time.sleep(1)


    browser.switch_to.default_content()
    browser.switch_to.frame('entryIframe')

    print(browser.find_element(By.CSS_SELECTOR,"span.LDgIH").text)
    browser.switch_to.default_content()
    time.sleep(1)
    
    browser.find_element(By.CSS_SELECTOR,".btn_clear").click()
    #     browser.execute_script("arguments[0].value = '';", search)     이 방법도 있긴함

```
