## 1-4. 지역별 병원명 추출

```python
city_name = []
town_name = []
hospital_name = []

for i in range(len(integratedCitiesData)) :
    params ={'serviceKey' : decoding,
             'STAGE1' : integratedCitiesData.iloc[i,0],
             'STAGE2' : integratedCitiesData.iloc[i,1],
             'pageNo' : '1',
             'numOfRows' : '10' }

    response = requests.get(url, params=params)
    dic = xmltodict.parse(response.content)
    
    city = integratedCitiesData.iloc[i,0]
    town = integratedCitiesData.iloc[i,1]

    
    print(f"{i+1}. {city} {town}")  # 시도명 출력    
    
    try :
        # item의 개수가 1개 이상인 경우 
        items = dic["response"]["body"]["items"]["item"]

        for j in range(len(items)):            
            print("\t"+items[j]["dutyName"])
            
            hospital_name.append(items[j]["dutyName"])
            city_name.append(city)
            town_name.append(town)

    except :
        try:
            # item의 개수가 1개인 경우
            item = dic["response"]["body"]["items"]["item"]
            print("\t"+item["dutyName"])
            
            hospital_name.append(item["dutyName"])
            city_name.append(city)
            town_name.append(town)
            
        except : 
            # item의 개수가 0개인 경우
            hospital_name.append(None)    
            city_name.append(city)
            town_name.append(town)
            print("\t NaN")
            
```

```python
pd.set_option('display.max_rows', 500)
```

```python
integratedCities_hospital = pd.DataFrame([city_name,town_name,hospital_name],index=["city","town","hospital"]).transpose()
integratedCities_hospital
```

```python
import warnings
warnings.filterwarnings("ignore")

# 네이버 검색 API에서 검색에 방해되는 단어 제거 
remove_lst = ["학교법인","의료법인","재단법인","을지학원","세종","가톨릭학원가톨릭대학교대전",
              "(재)미리내천주성삼성직수도회","한국보훈복지의료공단","고려중앙학원",
              "( 천주교부산교구유지재단)","(의)","()",")"]

for i in range(len(integratedCities_hospital["hospital"])):        
    modified_hospital = integratedCities_hospital["hospital"][i]
    for word in remove_lst : 
        modified_hospital = integratedCities_hospital["hospital"][i].replace(word,"")
        integratedCities_hospital["hospital"][i] =  modified_hospital
print(integratedCities_hospital["hospital"])
```

```python
# 재단이 들어가는 부분 나누기  ex) [**재단 ,**병원]
for i in range(len(integratedCities_hospital["hospital"])):    
    length = len(integratedCities_hospital["hospital"][i].split("재단"))  # 길이 1이면 재단 단어가 없고, 길이가 2이면 재단 단어 있음
    if length == 2 : 
        modified_hospital = integratedCities_hospital["hospital"][i].split("재단")[1]
    else : 
        modified_hospital = integratedCities_hospital["hospital"][i].split("재단")[0]
    integratedCities_hospital["hospital"][i] =  modified_hospital
integratedCities_hospital
```

## 2-1 실제 데이터 위치 불러기


```python
def getLocation(hospital) : 
    params = {
        'query': hospital,
        'display': '10',
        'start': '1',
        'sort': 'random',
    }
    
    response = requests.get('https://openapi.naver.com/v1/search/local.json', params=params, headers=headers)
    search = response.json()
    return search
```

```python
road_address = []
road_mapx = []
road_mapy = []

headers = {
    'X-Naver-Client-Id': '6Q9w3YC8r3qfKhf11tTY',
    'X-Naver-Client-Secret': 'gVhmq760LU',
}

for idx,hospital in enumerate(integratedCities_hospital["hospital"]):
    
    hospital = integratedCities_hospital["city"][idx]  +" "+ hospital
    search = getLocation(hospital)
    print(hospital)
    
    try :
        address = search["items"][0]["roadAddress"]
        mapx = float(search["items"][0]["mapx"][:3] +"." +  search["items"][0]["mapx"][3:]) # 경도
        mapy = float(search["items"][0]["mapy"][:2] +"." +  search["items"][0]["mapy"][2:]) # 위도
        
        road_address.append(address)
        road_mapy.append(mapy)
        road_mapx.append(mapx)
#         print(f"{hospital} - ", end="")    
#         print(f"{address}")
    
    except : 
            print("위치 데이터 없음")
            road_address.append(None)
            road_mapy.append(None)
            road_mapx.append(None)

    
```

```python
# 도로명,위도,경도로 임시 데이터 프레임 만들고 기존 df 데이터프레임에 concat 해준다. 
tmp = pd.DataFrame([road_address,road_mapx,road_mapy],index=["address","mapx","mapy"]).transpose()
new_df = pd.concat([integratedCities_hospital,tmp],axis=1)
new_df
```

```python
new_df.dropna(inplace=True)  # 결측치 제거 (때때로 데이터들 request 잘 안되는데, 결측치로 처리)
```

# 3. 데이터셋을 Folium으로 나타내기

```python
import folium
```

```python
emergency = pd.read_csv("emergency_data.csv")
emergency.isna().sum()  # 특정 병원 address는 나타나지 않지만, 위도/경도가 있으니 위치 출력에 지장없음
```

```python
m = folium.Map(location=[ 36.321655, 127.378953],zoom_start=7)

for idx,hospital in enumerate(emergency["hospital"]) : 
    popup_html = f'<span style="font-style:italic;">{emergency["address"][idx]}</span>'
    popup = folium.Popup(popup_html, max_width=600, max_height=600)
    
    icon = folium.Icon(
        color = "darkred",
        icon_color='red',
        icon='a-solid fa-heart-pulse',
        prefix='fa'
    )
    
    folium.Marker(
        [emergency["mapy"][idx],emergency["mapx"][idx]],
        popup=popup,
        tooltip=f"<b>{hospital}</b>",
        icon=icon
    ).add_to(m)
    
m.save('map.html')
```
