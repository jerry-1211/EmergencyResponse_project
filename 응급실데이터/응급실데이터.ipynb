## 1-3. 지역별 병원명 추출

```python
city_name = []
town_name = []
hospital_name = []

for i in range(len(new_df)) :
    params ={'serviceKey' : decoding,
             'STAGE1' : new_df.iloc[i,0],
             'STAGE2' : new_df.iloc[i,1],
             'pageNo' : '1',
             'numOfRows' : '10' }

    response = requests.get(url, params=params)
    dic = xmltodict.parse(response.content)
    
    city = new_df.iloc[i,0]
    town = new_df.iloc[i,1]

    
    print(f"{i+1}. {city} {town}")  # 시도명 출력    
    
    try :
        # item의 개수가 1개 이상인 경우 
        items = dic["response"]["body"]["items"]["item"]

        for j in range(len(items)):            
            print("\t"+items[j]["dutyName"])
            
            hospital_name.append(items[j]["dutyName"])
            city_name.append(city)
            town_name.append(town)

    except :
        try:
            # item의 개수가 1개인 경우
            item = dic["response"]["body"]["items"]["item"]
            print("\t"+item["dutyName"])
            
            hospital_name.append(item["dutyName"])
            city_name.append(city)
            town_name.append(town)
            
        except : 
            # item의 개수가 0개인 경우
            hospital_name.append(None)    
            city_name.append(city)
            town_name.append(town)
            print("\t NaN")
            
```

```python
df = pd.DataFrame([city_name,town_name,hospital_name],index=["city","town","hospital"]).transpose()
df
```

# 2. Naver search api

**목표** : <br>
- 네이버 검색 api를 가져오기
- 병원명만으로 위치(도로명,위도,경도) 추출 후 emergency.csv 파일에 저장

```python
road_address = []
road_mapx = []
road_mapy = []

headers = {
    'X-Naver-Client-Id': '6Q9w3YC8r3qfKhf11tTY',
    'X-Naver-Client-Secret': 'gVhmq760LU',
}

for hospital in df["hospital"] :    
    params = {
        'query': hospital,
        'display': '10',
        'start': '1',
        'sort': 'random',
    }

    print(f"[{hospital}] - ", end='   ' )
    
    response = requests.get('https://openapi.naver.com/v1/search/local.json', params=params, headers=headers)
    search = response.json()
    
    try :
        address = search["items"][0]["roadAddress"]
        mapx = float(search["items"][0]["mapx"][:3] +"." +  search["items"][0]["mapx"][3:]) # 경도
        mapy = float(search["items"][0]["mapy"][:2] +"." +  search["items"][0]["mapy"][2:]) # 위도
        
        road_address.append(address)
        road_mapy.append(mapy)
        road_mapx.append(mapx)
        
        print(f"{address} - ({mapy,mapy})")
    
    except : 
        print("위치 데이터 없음")
        road_address.append(None)
        road_mapy.append(None)
        road_mapx.append(None)
```

```python
# 도로명,위도,경도로 임시 데이터 프레임 만들고 기존 df 데이터프레임에 concat 해준다. 
tmp = pd.DataFrame([road_address,road_mapx,road_mapy],index=["address","mapx","mapy"]).transpose()
new_df = pd.concat([df,tmp],axis=1)
new_df
```

# 3. 데이터셋을 Folium으로 나타내기

```python
import folium
```

```python
emergency
```

```python
m = folium.Map(location=[ 36.321655, 127.378953],zoom_start=7)

for idx,hospital in enumerate(emergency["hospital"]) : 
    popup_html = f'<span style="font-style:italic;">{emergency["address"][idx]}</span>'
    popup = folium.Popup(popup_html, max_width=600, max_height=600)
    
    icon = folium.Icon(
        color = "darkred",
        icon_color='red',
        icon='a-solid fa-heart-pulse',
        prefix='fa'
    )
    
    folium.Marker(
        [emergency["mapy"][idx],emergency["mapx"][idx]],
        popup=popup,
        tooltip=f"<b>{hospital}</b>",
        icon=icon
    ).add_to(m)
m
```
